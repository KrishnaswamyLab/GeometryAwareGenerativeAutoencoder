```python
import pandas as pd
```

```python
res_df = pd.read_csv("affinity_synth_results.csv")
```

```python
res_df.sort_values(['seedmethod', 'bcv', 'dropout', 'probmethod'])
```

```python
res_df_merge = res_df.drop('accuracy', axis=1)

mean_df = res_df_merge.iloc[:,1:].groupby(['bcv', 'dropout', 'probmethod']).mean().round(3)
std_df = res_df_merge.iloc[:,1:].groupby(['bcv', 'dropout', 'probmethod']).std().round(3)

# Merge the mean and std dataframes
combined_df = mean_df.merge(std_df, left_index=True, right_index=True, suffixes=('_mean', '_std'))

# Format the results into a single column for each feature
for col in mean_df.columns:  # Iterate directly through the columns
    combined_df[col] = combined_df[col+'_mean'].astype(str) + "(±" + combined_df[col+'_std'].astype(str) + ")"
    # Since we're directly modifying the combined_df, no need to drop the columns in this loop

# Drop the original mean and std columns after processing
combined_df = combined_df.loc[:, ~combined_df.columns.str.contains('_mean|_std')]

# combined_df will now have the desired format
print(combined_df)
```

```python
res_df_dist = pd.read_csv("synth_results_w_recon.csv")
res_df1 = pd.read_csv("synth_results_many.csv")
```

```python
res_df_merge_dist = pd.concat([res_df_dist, res_df1['recon_weights']], axis=1)
```

```python
res_df_merge_dist['recon_weights'] = res_df_merge_dist['recon_weights'].apply(lambda x: (x == '[0.9, 0.1, 0]'))
res_df_merge_dist.rename(columns={'recon_weights': 'Reconstruction Loss'}, inplace=True)
```

```python
res_df_merge_dist.sort_values(['seedmethod', 'bcv', 'Reconstruction Loss'], inplace=True)
```

```python
cols = ['seedmethod', 'bcv', 'dropout', 'Reconstruction Loss', 'demap', 'accuracy', 'recon score']
res_df_merge_dist = res_df_merge_dist[cols]
```

```python
import pandas as pd

mean_df = res_df_merge_dist.iloc[:,1:].groupby(['bcv', 'dropout', 'Reconstruction Loss']).mean().round(3)
std_df = res_df_merge_dist.iloc[:,1:].groupby(['bcv', 'dropout', 'Reconstruction Loss']).std().round(3)

# Merge the mean and std dataframes
combined_df_dist = mean_df.merge(std_df, left_index=True, right_index=True, suffixes=('_mean', '_std'))

# Format the results into a single column for each feature
for col in mean_df.columns:  # Iterate directly through the columns
    combined_df_dist[col] = combined_df_dist[col+'_mean'].astype(str) + "(±" + combined_df_dist[col+'_std'].astype(str) + ")"
    # Since we're directly modifying the combined_df_dist, no need to drop the columns in this loop

# Drop the original mean and std columns after processing
combined_df_dist = combined_df_dist.loc[:, ~combined_df_dist.columns.str.contains('_mean|_std')]

# combined_df_dist will now have the desired format
print(combined_df_dist)
```
