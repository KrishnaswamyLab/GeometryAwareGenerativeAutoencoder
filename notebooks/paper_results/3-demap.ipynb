```python
import pandas as pd
```

```python
dimension = 10
plusminus = ''
res_dmae = pd.read_csv('../neurips_synth_results_w_recon.csv')
res_dmae.rename(columns={'seed': 'method', 'method': 'seed'}, inplace=True)
res_df = res_dmae[res_dmae['dimensions.latent'] == dimension]

res_df_dmae = res_df.copy()

mean_df = res_df.groupby(['method', 'bcv', 'dropout']).mean().round(3)
std_df = res_df.groupby(['method', 'bcv', 'dropout']).std().round(3)

# Merge the mean and std dataframes
combined_df = mean_df.merge(std_df, left_index=True, right_index=True, suffixes=('_mean', '_std'))

# Format the results into a single column for each feature
for col in mean_df.columns:  # Iterate directly through the columns
    combined_df[col] = combined_df[col+'_mean'].astype(str) + f" ({plusminus}" + combined_df[col+'_std'].astype(str) + ")"
    # Since we're directly modifying the combined_df, no need to drop the columns in this loop

# Drop the original mean and std columns after processing
combined_df_dmae = combined_df.loc[:, ~combined_df.columns.str.contains('_mean|_std')]
combined_df_dmae.drop(['seed', 'accuracy'], axis=1, inplace=True)

res_ae = pd.read_csv('../neurips_synth_results_w_recon_ae.csv')
res_ae.rename(columns={'seed': 'method', 'method': 'seed'}, inplace=True)
res_df = res_ae[res_ae['dimensions.latent'] == dimension]

res_df_ae = res_df.copy()

mean_df = res_df.groupby(['method', 'bcv', 'dropout']).mean().round(3)
std_df = res_df.groupby(['method', 'bcv', 'dropout']).std().round(3)

# Merge the mean and std dataframes
combined_df = mean_df.merge(std_df, left_index=True, right_index=True, suffixes=('_mean', '_std'))

# Format the results into a single column for each feature
for col in mean_df.columns:  # Iterate directly through the columns
    combined_df[col] = combined_df[col+'_mean'].astype(str) + f" ({plusminus}" + combined_df[col+'_std'].astype(str) + ")"
    # Since we're directly modifying the combined_df, no need to drop the columns in this loop

# Drop the original mean and std columns after processing
combined_df_ae = combined_df.loc[:, ~combined_df.columns.str.contains('_mean|_std')]
combined_df_ae.drop(['seed', 'accuracy'], axis=1, inplace=True)

res_no_exp = pd.read_csv('../neurips_synth_results_w_recon_no_exp.csv')
res_no_exp.rename(columns={'seed': 'method', 'method': 'seed'}, inplace=True)
res_df = res_no_exp[res_no_exp['dimensions.latent'] == dimension]

res_df_no_exp = res_df.copy()

mean_df = res_df.groupby(['method', 'bcv', 'dropout']).mean().round(3)
std_df = res_df.groupby(['method', 'bcv', 'dropout']).std().round(3)

# Merge the mean and std dataframes
combined_df = mean_df.merge(std_df, left_index=True, right_index=True, suffixes=('_mean', '_std'))

# Format the results into a single column for each feature
for col in mean_df.columns:  # Iterate directly through the columns
    combined_df[col] = combined_df[col+'_mean'].astype(str) + f" ({plusminus}" + combined_df[col+'_std'].astype(str) + ")"
    # Since we're directly modifying the combined_df, no need to drop the columns in this loop

# Drop the original mean and std columns after processing
combined_df_no_exp = combined_df.loc[:, ~combined_df.columns.str.contains('_mean|_std')]
combined_df_no_exp.drop(['seed', 'accuracy'], axis=1, inplace=True)
```

```python
import pandas as pd

def combine_and_compare(df1, df2):
    # Reset the index to convert MultiIndex to columns
    df1 = df1.reset_index()
    df2 = df2.reset_index()
    
    # Add a column to distinguish models
    df1['model'] = 'GAGA'
    df2['model'] = 'autoencoder'
    
    # Combine the two dataframes
    combined_df = pd.concat([df1, df2])
    
    # Round the bcv values
    combined_df['bcv'] = combined_df['bcv'].round(2).apply(str)
    
    # Split into GAGA and autoencoder DataFrames
    GAGA_df = combined_df[combined_df['model'] == 'GAGA'].drop(columns=['model'])
    autoencoder_df = combined_df[combined_df['model'] == 'autoencoder'].drop(columns=['model'])
    
    # Merge the DataFrames on method and bcv
    merged_df = pd.merge(GAGA_df, autoencoder_df, on=['method', 'bcv'], suffixes=('_GAGA', '_autoencoder'))
    
    # Create DataFrames for demap, DRS, and DGCS comparisons
    demap_df = merged_df[['method', 'bcv', 'demap_GAGA', 'demap_autoencoder', 'demap_phate_GAGA', 'demap_phate_autoencoder']].rename(columns={
        'demap_GAGA': 'GAGA', 
        'demap_autoencoder': 'autoencoder',
        'demap_phate_GAGA': 'phate',
        'demap_phate_autoencoder': 'autoencoder_phate'
    }).drop(columns=['autoencoder_phate'])
    drs_df = merged_df[['method', 'bcv', 'DRS_GAGA', 'DRS_autoencoder']].rename(columns={
        'DRS_GAGA': 'GAGA', 
        'DRS_autoencoder': 'autoencoder'
    })
    dgcs_df = merged_df[['method', 'bcv', 'DGCS_GAGA', 'DGCS_autoencoder']].rename(columns={
        'DGCS_GAGA': 'GAGA', 
        'DGCS_autoencoder': 'autoencoder'
    })
    
    return demap_df, drs_df, dgcs_df

def dataframe_to_latex(df, caption, label):
    # Convert DataFrame to LaTeX with custom formatting
    latex_str = df.to_latex(index=False, 
                            caption=caption, 
                            label=label, 
                            column_format='|l|c|c|c|c|c|',
                            longtable=False, 
                            escape=False)

    # Custom formatting for NeurIPS style
    formatted_latex = latex_str.replace('\\toprule', '\\hline')
    formatted_latex = formatted_latex.replace('\\midrule', '\\hline')
    formatted_latex = formatted_latex.replace('\\bottomrule', '\\hline')
    
    return formatted_latex

```

```python
demap_df, drs_df, dgcs_df = combine_and_compare(combined_df_dmae, combined_df_ae)
```

```python
demap_latex = dataframe_to_latex(demap_df, "Comparison of DEMaP values", "tab:demap_comparison")
drs_latex = dataframe_to_latex(drs_df, "Comparison of DRS values", "tab:drs_comparison")
dgcs_latex = dataframe_to_latex(dgcs_df, "Comparison of DGCS values", "tab:dgcs_comparison")

print(demap_latex)
print(drs_latex)
print(dgcs_latex)
```
