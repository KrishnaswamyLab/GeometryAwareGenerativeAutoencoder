```python
from plotly3d.plot import scatter, trajectories
import numpy as np
import pickle
```

```python
import pickle
with open(f'figure4_EB_geodesic.pkl', 'rb') as f:
    plot_dict = pickle.load(f)
```

```python
plot_dict.keys()
```

```python
colors = plot_dict['colors']
z_proc = plot_dict['z']
z0_proc = plot_dict['z0']
z1_proc = plot_dict['z1']
zhat_proc = plot_dict['traj']
```

```python
with open(f'figure4_EB_cross.pkl', 'rb') as f:
    cross_dict = pickle.load(f)
cross_idx = cross_dict['cross_idx']
cross_txt = cross_dict['cross_txt']
```

```python
fig = scatter(z_proc, colors, s=8)
fig.update_traces(marker=dict(
    line=dict(
        color='black',
        width=1
    )
))
# fig = scatter(z0_proc, s=10, fig=fig)
# fig = scatter(z1_proc, s=10, fig=fig)
# fig = scatter(z_proc[cross_idx], s=10, fig=fig)
for i in range(len(cross_idx)):
    ax=50
    ay=50
    if cross_txt[i] == 'NE-1':
        ay = 80
        ax = 80
    if cross_txt[i] == 'EN-2':
        ax = -80
        ay = -80
    if cross_txt[i] in ['NE/NC','EN']:
        ax = 100
        ay = 100
    xx, yy = z_proc[cross_idx].T
    fig.add_annotation(
        x=xx[i],
        y=yy[i],
        text=cross_txt[i],
        showarrow=True,
        arrowhead=2,     # Choose the arrowhead style
        arrowsize=1,     # Adjust the size of the arrowhead
        arrowwidth=2,    # Adjust the width of the arrow
        ax=ax,           # Positive value to move arrow start point to the right
        ay=ay,           # Positive value to move arrow start point downward
        arrowcolor='black',  # Color of the arrow
        font=dict(size=20, color='black'), 
        xshift=0,  # Adjust the xshift value as needed
        yshift=0   # Adjust the yshift value as needed
    )

fig = trajectories(zhat_proc, s=5, fig=fig, title='', white_bkgrnd=True, ticks=False, figsize=(10,4))
fig.update_layout(
    title='',
    showlegend=False
)
fig.show()
fig.write_image("eb_geodesic.pdf", format='pdf')
```
