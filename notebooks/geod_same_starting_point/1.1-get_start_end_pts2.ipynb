```python
import numpy as np
from plotly3d.plot import scatter
import pathlib
```

```python
data_path = '../../data/neurips_results/toy/gt'
file_extn = '_none_0.npz'
data_names = ['hemisphere', 'ellipsoid', 'saddle', 'torus']
```

```python
data_dict = {}
for data_name in data_names:
    data = np.load(f"{data_path}/{data_name}{file_extn}")
    points = data['X']
    if data_name != 'saddle':
        scores = points[:, 0] + points[:, 1] + 2*points[:, 2]
        sorted_indices = np.argsort(scores)[::-1]
        num_front_points = int(len(points) * 0.2)
    else:
        scores = -0.5*points[:, 0] + 0.5*points[:, 1] + points[:, 2]
        sorted_indices = np.argsort(scores)[::-1]
        num_front_points = int(len(points) * 0.3)
    front_indices = sorted_indices[:num_front_points]
    np.random.seed(3243)
    start_idx = np.random.choice(front_indices, 1, replace=False)
    n_points = 10
    end_idx = np.random.choice(front_indices, n_points, replace=False)
    start_points = points[start_idx]
    start_points = np.tile(start_points, (n_points,1))
    end_points = points[end_idx]

    data_dict[data_name] = dict(
        X=points,
        X_ground_truth=points,
        start_points=start_points,
        end_points=end_points
    )

    fig = scatter(points)
    fig = scatter(start_points, fig=fig, s=3)
    fig = scatter(end_points, fig=fig, s=3, title=data_name)
    fig.show()
```

```python
pathlib.Path('../../data/neurips_results/toy/visualize2/').mkdir(parents=True, exist_ok=True)
```

```python
for data_name in data_names:
    np.savez(f'../../data/neurips_results/toy/visualize2/{data_name}.npz', **data_dict[data_name])
```
